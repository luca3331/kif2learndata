```python
import os
```

```python
def read_txtfile(folder_path='kif_txt'):
    extension = '.txt'    # 開くファイルの拡張子を指定

    # フォルダ内の対象の拡張子を持つファイル一覧を取得
    file_list = [f for f in os.listdir(folder_path) if f.endswith(extension) and os.path.isfile(os.path.join(folder_path, f))]
    files = []
    for file_name in file_list:
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r') as file:
            lines = file.readlines()
            files.append((file_name, lines))
    return files
```

```python
folder_path = 'kif_txt'
files = read_txtfile(folder_path)
```

```python
class Kif:
    def __init__(self, file):
        self.name = file[0]
        self.kifs = str(file[1:]).split(',')
        self.tumos = str(file[-1]).split(',')[30]
        self.hands_len = str(file[-1]).split(',')[23]
        print(self.kifs)
        self.p1_fields_history, self.p2_fields_history = self.kif2history()
    
    def kif2history(self):
        p1_turn, p2_turn = 0, 0
        p1_history, p2_history = [self.kifs[0][13]], [self.kifs[0][13]]
        for turn_n in range(self.hands_len):
            kif = self.kifs[turn_n][1:]
            if kif[1] == '2':
                p1_history.append(kif[12])
                p2_history.append(kif[52])
            else:
                if kif[12] == p1_history[-1]:
                    p2_history.append(kif[52])
                else:
                    p1_history.append(kif[12])
        return p1_history, p2_history
    
```
